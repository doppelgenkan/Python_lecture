## 肘関節に関する力と力のモーメントのプログラム

Numpyを使って、下図のような上肢のモデルのシミュレーションをおこなうPython関数をプログラムしよう！プログラムに必要なNumpyモジュールは、以下のリンク先のページを開いてください：

<a href="./docs/appendix_numpy.html"><b>Numpyの簡単公式</b></a>

### ◾️パラメータ等
【調整可能パラメータ】 
- 体質量 : $w\ (\mathrm{kg})$
- 錘質量 : $M\ (\mathrm{kg})$
- 肘関節Oから錘の重心Qまでの距離 : OQ = $l\ (\mathrm{m})$  
- 肩関節屈曲角 : $\beta\ (\mathrm{degrees})$
- 肘関節屈曲角 : $\theta\ (\mathrm{degrees})\ (\theta>0)$
- 角条件 : $\beta\geq0^\circ,\ 10^\circ\leq\theta\leq135^\circ,\ \beta+\theta<180^\circ$

【与えられたまたは計算可能なパラメータ】 
- 前腕質量と手部の質量 : $m = 0.025w\ (\mathrm{kg})$
- 肘関節Oから前腕重心Gまでの距離 : OG = $h = 0.5l\ (\mathrm{m})$
- 肘関節Oから上腕筋の腱の装着部までの距離 : OP = $d=0.125l\ (\mathrm{m})$
- 重力加速度 : $g=9.8\ (\mathrm{m/s^2})$

【その他】
- 上腕筋の力$\bf{F}$は上腕と平行  

### ◾️計算式等

#### 1.  肘のまわりの外力のモーメントのz成分  

$M_{\mathrm{ex}} = -g(mh+Ml)\sin{(\beta+\theta)}$  

#### 2.  肘のまわりの上腕筋の力のモーメントのz成分

$M_z := M_{\mathrm{in}} = -M_{\mathrm{ex}}$

#### 3. 上腕筋の力の大きさ  

$F = -\frac{M_{\mathrm{ex}}}{d\sin{\theta}}=\frac{M_{\mathrm{in}}}{d\sin{\theta}}=\frac{M_{z}}{d\sin{\theta}}$

#### 4. 上腕筋の力のベクトルの成分

$F_{x} = -F\sin{\beta}$  
$F_{y} = F\cos{\beta}$

#### 5. 肘関節の抗力
$R_{x} = -F_{x}$  
$R_{y} = (m+M)g - F_{y}$



```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure, plot, xlabel, ylabel
%matplotlib inline

g = 9.8

def human_par(w, l): #wは体質量，lは前腕の長さ
    m = w * 0.025 #前腕と手の質量
    d = l * 0.125 #肘関節からの上腕筋の腱が付いている場所の距離
    h = l * 0.5 #肘関節から前腕と手の重心までの距離
    return np.array([m, d, h, l])

def M_z(par, load, beta, theta):
    m = par[0]
    h = par[2]
    l = par[3]
    M = load
    return g * (m*h + M*l) * np.sin(np.radians(beta) + np.radians(theta))

def F(par, load, beta, theta):
    d = par[1]
    return M_z(par, load, beta, theta) / (d * np.sin(np.radians(theta)))

def F_x(par, load, beta, theta):
    return -F(par, load, beta, theta) * np.sin(np.radians(beta))

def F_y(par, load, beta, theta):
    return F(par, load, beta, theta) * np.cos(np.radians(beta))

def R_x(par, load, beta, theta):
    return -F_x(par, load, beta, theta)

def R_y(par, load, beta, theta):
    m = par[0]
    M = load
    return (m + M)*g - F_y(par, load, beta, theta)

def R(par, load, beta, theta):
    return np.sqrt( R_x(par, load, beta, theta)**2 + R_y(par, load, beta, theta)**2 )

def theta_list(x):
    return np.arange(10, x+1, 1)
                
def do_calc():
    w = float(input('体質量(kg)を入力してください: '))
    l = float(input('前腕長(m)を入力してください : '))
    load = float(input('錘質量(kg)を入力してください: '))
    beta = float(input('肩屈曲角(degrees)を入力してください: '))
    theta = float(input('肘屈曲角(degrees)を入力してください: '))
    if beta < 0 or theta < 10 or theta > 135 or (beta + theta) >= 180:
        print('計算ストップ：不正な角です')
    else:
        my_par = human_par(w, l)
        mom = M_z(my_par, load, beta, theta)
        f = F(my_par, load, beta, theta)
        fx = F_x(my_par, load, beta, theta)
        fy = F_y(my_par, load, beta, theta)
        rx = R_x(my_par, load, beta, theta)
        ry = R_y(my_par, load, beta, theta)
        r = R(my_par, load, beta, theta)
        vec_f = np.array([fx, fy])
        vec_r = np.array([rx, ry])
        print('------------------')
        print('【計算結果】')
        print('肘関節モーメント(z成分)の値(Nm): %s' % mom)
        print('上腕筋の力の大きさ(N)         : %s' % f)
        print('肘関節抗力の大きさ(N)         : %s' % r)
        print('上腕筋の力のベクトル(N)        : %s' % vec_f)
        print('肘関節抗力のベクトル(N)        : %s' % vec_r)

def draw_effects():
    w = float(input('体質量(kg)を入力してください: '))
    l = float(input('前腕長(m)を入力してください : '))
    load = float(input('錘質量(kg)を入力してください: '))
    beta = float(input('肩屈曲角(degrees)を入力してください: '))
    my_par = human_par(w, l)
    if beta < 45:
        angles = theta_list(135)
    else:
        angles = theta_list(179 - beta)
    #肘関節モーメントのグラフ
    plt.figure()
    plt.plot(angles, M_z(my_par, load, beta, angles))
    plt.xlabel('Flection angles of elbow (degrees)')
    plt.ylabel('Moment of force around elbow (Nm)')
    #前腕筋力のグラフ
    plt.figure()
    plt.plot(angles, F(my_par, load, beta, angles))
    plt.xlabel('Flection angles of elbow (degrees)')
    plt.ylabel('Force of forearm mascle (N)')
    #肘関節筋力のグラフ
    plt.figure()
    plt.plot(angles, R(my_par, load, beta, angles))
    plt.xlabel('Flection angles of elbow (degrees)')
    plt.ylabel('Resistance of elbow (N)')
```

```python
do_calc()
```

```python
draw_effects()
```
