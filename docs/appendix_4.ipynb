### ４章の補足公式

***
**＜公式4-1＞**  
```
def ブロック名():
    処理1
    処理2
    処理3
    ....
    処理n
```
意味：`処理1`〜`処理n`**の一連の(ブロック)処理をまとめて定義します。この一連の処理を呼び出すには、セルに**  
```
ブロック名()
```  

**と入力し、[Shift]+[Enter]を押します。**  

> ＜例＞   
> ```
> In [1]: def proc1(): 
>             print('一回目の処理')
>             print('２回目のしょり')
>             print('さん回目のショリ')
> ```  
> を読み込み、   
> ```
> In [2]: proc1()
> ```  
> を実行する(呼び出す)と、  
> ```
>         一回目の処理
>         ２回目のしょり
>         さん回目のショリ
> ```  
> が表示されます。

- 実際には、この公式は**引数なし**の**関数定義**ですが、ここではあえてブロックとよぶことにします。  
- `def ブロック名():`の下のからブロックが終わるまでの行は、**半角4文字分のインデント(4文字の半角空白)**を入れなければなりません。  
- ブロックの**初めの文字(右側の文字)は同じインデント**で揃えなければなりません。
***

***  
**＜公式4-2a＞**  
```
変数 = input('文字列')  
```
意味：**指定の`文字列`の入力プロンプトが表示されます。プロンプトから入力した値は、(強制的に)文字列に変換されて指定の`変数`に代入されます。**  
> ＜例＞  
> ```  
> In [3]: val1 = input('何かを入力してください：')
> ```
> と入力し、**［Shift］＋［Enter］**を押すと、すぐ下に
> ```
>         氏名を入力してください：3.14  
> ```
> のように、プロンプトが表示されます。`3.14`を代入し**［Enter］**を押します。
> ```
> In [4]: val1 
> ```
> 上のように新しいセルに、`val1`を入力し**［Shift］＋［Enter］**を押すと、
> ```
> Out[4]: '3.14'  
> ```
> というように、`val1`の内容の文字列が表示されます。

***

***  
**＜公式4-2b＞**  
```
変数 = int(input('文字列'))
```

意味：**指定の`'文字列'`の入力プロンプトが表示されます。プロンプトから入力した整数または整数文字列が、(強制的に)整数に変換されて指定(左辺)の`変数`に代入されます。**  
> ＜例＞  
> ```  
> In [5]: val2 = int(input('数値を入力してください：')) 
> ```
> と入力し、**［Shift］＋［Enter］**を押すと、すぐ下に
> ```
>         数値を入力してください：137 
> ```
> のように、プロンプトが表示されます。`137`を代入し**［Enter］**を押します。
> ```
> In [6]: val2
> ```
> 上のように新しいセルに、`val2`を入力し**［Shift］＋［Enter］**を押すと、
> ```
> Out[6]: 137  
> ```
> というように、`val2`の内容の整数が表示されます。

- 関数 `int(引数)` は、指定の`引数`の**整数文字列**または**数値**(**整数**または**浮動小数**)を**整数**にして返します(引数が浮動小数の場合、**小数第１位以下を切り捨て**ます)。  
> 【補足】プロンプト表示後の数値入力で**浮動小数を入力した場合、エラー**を起こします。これを回避するためには、  
>
> ```
> 変数 = int(float(input('文字列')))
> ```
>
> としてください(**公式3-2c参照**)。

***

***  
**＜公式4-2c＞**  
```
変数 = float(input('文字列'))  
```
意味：**指定の`'文字列'`の入力プロンプトが表示されます。プロンプトから入力した整数、浮動小数または数字文字列が、(強制的に)浮動小数に変換されて指定(左辺)の`変数`に代入されます。**  
> ＜例＞  
> ```  
> In [7]: val3 = float(input('数値を入力してください：'))
> ```
> と入力し、**［Shift］＋［Enter］**を押すと、すぐ下に
> ```
>         数値を入力してください：3.14  
> ```
> のように、プロンプトが表示されます。`3.14`を代入し**［Enter］**を押します。 
> ```
> In [8]: val3
> ```
> 上のように新しいセルに、`val3`を入力し**［Shift］＋［Enter］**を押すと、
> ```
> Out[8]: 3.14  
> ```
> というように、`val3`の内容の浮動小数が表示されます。

- 関数`float(引数)`は、指定の`引数`の**数字文字列**または**数値**(**整数**または**浮動小数**)を**浮動小数**にして返します。
***
